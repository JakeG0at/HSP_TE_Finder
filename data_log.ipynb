```python
#packages
from __future__ import print_function
from auth import spreadsheet_service
from auth import drive_service
import pandas as pd
import numpy as np
import os
import glob
import yfinance as yf
import shutil
```

```python

# use glob to get all the csv files 
# in the folder
os.chdir('/home/jake/Documents/tastyworks_dailydata/')
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))
  
# loop over the list of csv files
for f in csv_files:
      
    # read the csv file
    df = pd.read_csv(f)
      
    # print the location and filename
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])
      
    # print the content
    print('Content:')
    display(df)
    print()
```

```python
#Remove the d from dte
df.DTE = df.DTE.str.rstrip("d")
#Remove the commas from the whole df
df.replace(',','', regex=True, inplace=True)
#Convert all the collumns to floats
df=df.astype({'Symbol':str,
 'Quantity':float,
 'DTE':float,
 'Strike Price':float,
 'Call/Put':str,
 'P/L Open':float, 
 'P/L Day':float,
 'IV Rank':float,
 'Theta':float,
 'Vega':float,
 'Delta':float})
#Round all the decimals
df=df.round(decimals =0)
#Remove the deltas that are --
#Remove the DTEs that are not >0
```

```python
#Short Puts     DTE>5 put, quuantity should be > -1
sp= df.loc[(df['DTE']>5) & (df['Quantity']< -1) & (df['Call/Put']=='Put')]
sp_unique=sp.count()
sp_quantity=sp.Quantity.sum()
sp_strike_Price=sp['Strike Price'].mean()
sp_dte=sp.DTE.mean()
sp_delta=sp.Delta.sum()
sp_theta=sp.Theta.sum()
sp_vega=sp.Vega.sum()
```

```python
#Roaming Short Puts  quanity= -1 put, 
rsp= df.loc[(df['Quantity']== -1) & (df['Call/Put']=='Put')]
rsp_unique=rsp.count()
rsp_quantity=rsp.Quantity.sum()
rsp_strike_Price=rsp['Strike Price'].mean()
rsp_dte=rsp.DTE.mean()
rsp_delta=rsp.Delta.sum()
rsp_theta=rsp.Theta.sum()
rsp_vega=rsp.Vega.sum()
```

```python
#Short Calls   DTE >5 quantity, (-), call, not a credit spread
sc=df.loc[(df['DTE']>5) & (df['Quantity']>0) & (df['Call/Put']=='Call')]
sc_unique=sc.count()
sc_quantity=sc.Quantity.sum()
sc_strike_Price=sc['Strike Price'].mean()
sc_dte=sc.DTE.mean()
sc_delta=sc.Delta.sum()
sc_theta=sc.Theta.sum()
sc_vega=sc.Vega.sum()
```
